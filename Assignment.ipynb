```python
import numpy as np
import pandas as pd
import matplotlib as mat
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#assignment 1- 
#Plot a graph for 
#x = np.range(0,10)
#y = x*x
```

```python
x=np.arange(0,10)
y=x*x
plt.plot(x, y, 'go--',linewidth=2, markersize=10)
plt.xlabel('x axis', color='blue')
plt.ylabel('y axis', color='blue')
plt.title('2D-Diagram', color='red')
plt.show()
```
